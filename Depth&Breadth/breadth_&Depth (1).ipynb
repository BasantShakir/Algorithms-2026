```python

graph = {
    'u': ['v', 'x'],
    'v': ['y'],
    'w': ['y', 'z'],
    'x': ['v'],
    'y': ['x'],
    'z': ['z']
}


color = {}  # WHITE, GRAY, BLACK
pi    = {}
d     = {}
f     = {}
time  = 0


WHITE, GRAY, BLACK = 0, 1, 2


for u in graph:
    color[u] = WHITE
    pi[u] = None

def DFS():
    global time
    time = 0
    for u in graph:
        if color[u] == WHITE:
            DFS_VISIT(u)


def DFS_VISIT(u):
    global time
    time += 1
    d[u] = time
    color[u] = GRAY

    for v in graph[u]:
        if color[v] == WHITE:
            pi[v] = u
            DFS_VISIT(v)

    color[u] = BLACK
    time += 1
    f[u] = time


DFS()


print("ta\td\tf")
for u in sorted(graph):
    print(f"{u}\t\t{d[u]}\t{f[u]}")
```

```python
from collections import deque

def BFS(G, s):
    color = {}
    d = {}
    π = {}

    # Initialize all vertices
    for node in G:
        color[node] = "WHITE"
        d[node] = float('inf')
        π[node] = None

    # Initialize source vertex
    color[s] = "GRAY"
    d[s] = 0
    π[s] = None

    Q = deque()
    Q.append(s)

    while Q:
        u = Q.popleft()

        for v in G[u]:
            if color[v] == "WHITE":
                color[v] = "GRAY"
                d[v] = d[u] + 1
                π[v] = u
                Q.append(v)

        color[u] = "BLACK"
        print(f"Visited: {u}, Distance: {d[u]}, Parent: {π[u]}")

    return d, π

# Graph
G = {
    's': ['w', 'r'],
    'w': ['t', 'x'],
    'r': ['v'],
    'v': [],
    't': ['u'],
    'x': ['y'],
    'u': [],
    'y': []
}

# Run BFS
d, π = BFS(G, 's')

print(f"\nFinal Results:")
for node in G:
    print(f"{node}: Distance = {d[node]}, Parent = {π[node]}")



```
